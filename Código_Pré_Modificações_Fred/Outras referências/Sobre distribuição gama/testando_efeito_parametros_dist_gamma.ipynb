```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Configurações
MWm = 700
x = np.linspace(700, 4000, 1000)

# Valores de alfa para comparar
alphas = [1.5, 2.5, 5.0]
beta = 300

plt.figure(figsize=(8, 5))
for alpha in alphas:
    y = gamma.pdf(x - MWm, a=alpha, scale=beta)
    plt.plot(x, y, label=f'alfa = {alpha}, β = {beta}')

plt.title("Efeito de alfa (forma) na distribuição gama")
plt.xlabel("Massa molar (g/mol)")
plt.ylabel("Frequência relativa")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
